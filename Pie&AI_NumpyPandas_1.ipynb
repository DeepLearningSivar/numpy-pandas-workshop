# Listas tradicionales vs Numpy Arrays

```python
weights = [65.4,78.8,80.5]
heights = [1.70,1.65,1.95]

print('Persona 1: ',weights[0],heights[0])
```

```python
#IMC
# Formular = pesor / altura **2
```

```python
np_weights = np.array(weights)
np_heights = np.array(heights)
imc = np_weights / np_heights ** 2
print(imc)
```

```python
listExample = [1,2,3]
np_example = np.array([1,2,3])

print('Lista',listExample,type(listExample))
print('NP Array',np_example, type(np_example))
```

```python
print(listExample+listExample)
print(np_example+np_example)
```

```python
imc
```

```python
n_imc = np.round(imc, 2)
print(n_imc)
```

```python
print(n_imc > 22)
```

```python
n_imc[n_imc > 22]
```
